# Linear Regression for Passing Data

Age = x1, Attempts = x2, Completions = x3, Yards = x4, Interceptions = x5, Yards per Attempt = x6, Yards per Game = x7, Touchdowns = y

## Full Model

```python
#using the columns of Age = x1,  Attempts = x2, Completions = x3, Yards = x4, Interceptions = x5, Yards per Attempt = x6, Yards per Game = x7, Touchdowns = y 
x = passing_data[['Age','Att', 'Cmp', 'Yds','Int', 'Y/A', 'Y/G']]
y = passing_data[['TD']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)

from sklearn.linear_model import LinearRegression
full = LinearRegression()
full.fit(x_train, y_train)
```

```python
for idx, col_name in enumerate(x_train.columns):
    print("The coefficient for {} is {}".format(col_name, full.coef_[0][idx]))
    
intercept = full.intercept_[0]
print("The intercept for our model is {}".format(intercept))
```

The full regression model is y = .02046 + .017122x1 - .0424x2 + .0185x3 + .011744x4  - .21396795x5 - .0569772x6 - .00603x7

```python
print("The R^2 value is {}".format(full.score(x_test, y_test)))
```

This means that 91.66% of the variability in Age, Attempts, Completions, Yards, Interceptions, Yards per Attempt, and Yards per Game
